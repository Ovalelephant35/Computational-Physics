Plot of a single variable function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values for the plot
x = np.linspace(-10, 10, 400)  # Adjust the range as needed

# Define the mathematical expression
expression = "x**3 + 2*x + 1"

# Evaluate the expression for the given x values
try:
    y = eval(expression)
    
    # Create a plot
    plt.plot(x, y, label=expression)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Plot of {expression}')
    plt.legend()
    plt.grid(True)
    plt.show()

except (SyntaxError, NameError) as e:
    print("Invalid input. Please enter a valid mathematical expression.")
```

Plot for Two variable function

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x and y values
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)

# Create a grid of (x, y) values
X, Y = np.meshgrid(x, y)

# Define the two-variable function (e.g., f(x, y) = x^2 + y^2)
# Replace this with your own function
Z = X**2 + Y**2

# Create a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour, label='Function Value')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of a Two-Variable Function')
plt.grid(True)
plt.show()
```

Multiple Function on same graph

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-5, 5, 400)

# Define multiple functions (you can add more if needed)
function1 = x**2
function2 = np.sin(x)
function3 = np.cos(x)

# Create a plot
plt.figure(figsize=(8, 6))

# Plot each function with a label
plt.plot(x, function1, label='x^2', color='blue')
plt.plot(x, function2, label='sin(x)', color='red')
plt.plot(x, function3, label='cos(x)', color='green')

# Customize the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Functions on the Same Graph')
plt.legend()  # Show legend with labels
plt.grid(True)

# Show the plot
plt.show()
```
