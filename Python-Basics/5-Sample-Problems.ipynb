```python
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0,2,100)
g=9.81
vj=10
vm=15
yj= -0.5*g*t**2 + vj*t
ym= -0.5*g*t**2 + vm*t
plt.plot(t, yj , label="jim")
plt.plot(t,ym , label="mark")
plt.xlabel('time')
plt.ylabel('Height')
plt.legend()
```
