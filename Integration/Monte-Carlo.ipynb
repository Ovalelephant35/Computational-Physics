```python
import scipy.integrate as spi
import numpy as np

# Define the radius of the sphere
R = 3.0  # Replace with the desired radius

# Define the integrand
def integrand(x):
    return np.pi * (R**2 - x**2)

# Perform the integration
volume, _ = spi.quad(integrand, -R, R)

# Print the volume of the sphere
print(f"Volume of the sphere with radius {R} is approximately {volume:.6f}")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
a = 0
b = np.pi
N =10000
M=1000
num_reps=M
num_simulations=N
def f(x):
  return np.exp(x)
plt_vals =[]
for i in range(num_simulations):
  ar = np.zeros(num_reps)
  for i in range (len(ar)): 
    ar[i] = np.random.uniform(a,b)
  integral = 0.0
  for i in ar: 
    integral += f(i)
  ans = (b-a)/float(num_reps)*integral 
  plt_vals.append(ans) 
plt.title("Distributions of areas calculated") 
plt.hist (plt_vals, bins=30, ec="black")  
plt.xlabel("Areas") 
plt.show()
print(plt_vals)
```
