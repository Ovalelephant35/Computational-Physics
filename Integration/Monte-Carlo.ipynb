```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to be integrated
def f(x):
    return x**3 * np.exp(-x**2)

# Define the integration limits
a = 0.0  # Lower limit
b = 2.0  # Upper limit

# Number of random samples
n = 10000

# Generate random x values within the integration limits
x_values = np.random.uniform(a, b, n)

# Calculate the function values at the random x values
y_values = f(x_values)

# Calculate the Monte Carlo estimate of the integral
integral_estimate = (b - a) * np.mean(y_values)

# Create a histogram to visualize the distribution of random points
plt.hist(x_values, bins=50, density=True, alpha=0.5, label='Random Samples')
x_range = np.linspace(a, b, 1000)
plt.plot(x_range, f(x_range), 'r-', label='Function f(x)')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.title('Monte Carlo Integration')
plt.legend()
plt.grid(True)

# Print the integral estimate
print(f"Monte Carlo Integration Estimate: {integral_estimate:.6f}")

plt.show()

```

```python
import numpy as np 
import matplotlib.pyplot as plt 
a = 0
b = np.pi
N = 10000
M=1000
num_reps=M
num_simulations=N
def f(x):
  return np.sin(x)
plt_vals =[]
for i in range(num_simulations):
  ar = np.zeros(num_reps)
  for i in range (len(ar)): 
    ar[i] = np.random.uniform(a,b)
  integral = 0.0
  for i in ar: 
    integral += f(i)
  ans = (b-a)/float(num_reps)*integral 
  plt_vals.append(ans) 
plt.title("Distributions of areas calculated") 
plt.hist (plt_vals, bins=30, ec="black")  
plt.xlabel("Areas") 
plt.show()
print(plt_vals)
```
